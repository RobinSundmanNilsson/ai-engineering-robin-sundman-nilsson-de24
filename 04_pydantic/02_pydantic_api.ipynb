### Using pydantic to de-serialize the json object into pydantic model

```python
data["results"][:10]
```

```python
from pydantic import BaseModel

class PokemonListResponse(BaseModel):
    count: int
    results: list[dict]

pokemons = PokemonListResponse.model_validate(data)
pokemons
```

```python
type(pokemons)
```

```python
pokemons.count
```

```python
pokemons.results
```

```python
from pydantic import ValidationError

class PokemonListResponse(BaseModel):
    count: str
    results: list[dict]

try:
    pokemons = PokemonListResponse.model_validate(data)
except ValidationError as e:
    print("Validation error:", e)
```

```python
data = requests.get("https://pokeapi.co/api/v2/pokemon-species/pikachu").json()
data.keys()
```

```python
for key, values in data.items():
    print(f"{key}: {type(values)}")
```

```python
class Pokemon(BaseModel):
    id: int
    name: str
    base_happiness: int
    capture_rate: int
    is_legendary: bool
    gender_rate: int
    evolution_chain: dict
    generation: dict

pikachu = Pokemon.model_validate(data)
pikachu
```

```python
pikachu.evolution_chain
```

```python
pikachu.generation
```

```python
pikachu.base_happiness
```

```python
pikachu.model_dump_json()
```

```python
serialized_pikachu = pikachu.model_dump_json()
```
